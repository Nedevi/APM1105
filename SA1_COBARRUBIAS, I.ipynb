```python
import requests
import re

#1

url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'

response = requests.get(url)

if response.status_code == 200:
    with open('words.txt', 'wb') as f:
        f.write(response.content)
        print('Dataset saved to file.')
else:
    print('Failed to download dataset.')

#A
words = []
count = 0

with open ("words.txt","r") as f:
    words = set(f.readlines()) 
    count = len(words) 

print("1a.", count)
#/A
#B
def get_longest_word(a):
    with open("words.txt", "r") as f:
        content = f.read()
    words = re.findall(r'\b\w+\b', content)
    longest_word = max(words, key=len)
    return longest_word
    
word = get_longest_word("words.txt")
print("1b.", word)

#/B
#C
def count_words(file, s):
    with open(file, "r") as f:
        content = f.read()
    words = re.findall(r'\b\w+\b', content)
    count = 0
    for word in words:
        if word.startswith(s):
            count += 1
    return count
q = str(input("Input letter:"))
print("1c.", count_words('words.txt', q))
        
#/C
#D
def words_substring(file_path, substring):
    with open(file_path, "r") as f:
        content = f.read()
    words = re.findall(r'\b\w+\b', content)
    words_substring = [word for word in words if substring in word]
    return words_substring

a = input("Input substring: ")
print("1d. Words containing", a, words_substring('words.txt', a))

#/D
#E
def is_palindrome(word):
    cleaned_word = re.sub(r'[^a-zA-Z0-9]', '', word.lower())
    return cleaned_word == cleaned_word[::-1]

def palindromes(file_path):
    with open(file_path, "r") as f:
        content = f.read()
    words = re.findall(r'\b\w+\b', content)
    palindromes = set()
    for word in words:
        if is_palindrome(word):
            palindromes.add(word)
    return palindromes

print("\n1e. Palindromes:", palindromes('words.txt'))

#/E
#F
def count_letters(input_string):
    input_string = input_string.lower()
    letter_counts = {}
    for char in input_string:
        if char.isalpha():
            letter_counts[char] = letter_counts.get(char, 0) + 1
    return letter_counts

with open("words.txt", "r") as f:
    words = f.read().split()

letter_counts = {}
for word in words:
    word_counts = count_letters(word)
    for char, count in word_counts.items():
        letter_counts[char] = letter_counts.get(char, 0) + count

print("Letter counts:", letter_counts)
#/F
```

```python
import time


def fibonacci_iter(n):
    if n <= 0:
        return "Invalid input. Please enter a positive integer."
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        a, b = 0, 1
        for _ in range(2, n):
            a, b = b, a + b
        return b

def fibonacci_recursion(n):
    if n <= 0:
        return "Invalid input. Please enter a positive integer."
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibonacci_recursion(n - 1) + fibonacci_recursion(n - 2)


start_time_iter = time.time()
result_iter = fibonacci_iter(30) 
end_time_iter = time.time()
time_taken_iter = end_time_iter - start_time_iter


start_time_recursion = time.time()
result_recursion = fibonacci_recursion(30) 
end_time_recursion = time.time()
time_taken_recursion = end_time_recursion - start_time_recursion

print("Fibonacci using iteration (30th number):", result_iter)
print("Time taken by iteration:", time_taken_iter, "seconds")
print()
print("Fibonacci using recursion (30th number):", result_recursion)
print("Time taken by recursion:", time_taken_recursion, "seconds")
```

```python
def cubic_root_iteration(x0, max_iterations=1000, tolerance=1e-6):
    x = x0
    for i in range(max_iterations):
        x_next = pow(20 - 5 * x, 1/3)
        if abs(x_next - x) < tolerance:
            return round(x_next, 3)
        x = x_next
    return round(x, 3)

x0 = 2

solution = cubic_root_iteration(x0)

print(f"x0: {x0:.3f}")
print(f"x1: {solution:.3f}")
```
