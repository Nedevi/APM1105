```python
import pandas as pd
import numpy as n
import matplotlib.pyplot as m
import seaborn as s
```

```python
file_path = '/Users/imee/Downloads/Spotify_Youtube.csv'
df = pd.read_csv(file_path)
df = df.drop_duplicates(keep = 'first')
df.head(10)
```

```python
df.columns
```

```python
print("Dataset Info:")
print(df.info())

print("\nColumn Counts:")
print(df.count())
print("\nNumber of Data Points:", df.shape[0])

null_columns = df.columns[df.isnull().any()]
print("\nColumns with Null Values:")
print(df[null_columns].isnull().sum())

print("\nData Types Counts:")
print(df.dtypes.value_counts())


df['Likes'].fillna(0, inplace=True)
df['Comments'].fillna(0, inplace=True)

df.dropna(inplace=True)

print("\nUpdated Dataset Info:")
df.info()
```

```python
df.iloc[:, 1:].describe().round(2)
```

```python
album_counts = df['Album_type'].value_counts()
album_counts
```

```python
m.figure(figsize=(6, 6))
m.pie(album_counts, labels=album_counts.index, autopct='%1.1f%%')
m.title('Distribution of Album Types')
m.show()
```

## Data shows that the tracks in the dataset are mostly from album releases, taking up nearly 3/4ths. 24.1% comes from singles and only 3.6% are compilations.

```python
youtube_data = df[['Artist', 'Views']].groupby('Artist').sum().nlargest(10, 'Views')
spotify_data = df[['Artist', 'Stream']].groupby('Artist').sum().nlargest(10, 'Stream')

m.figure(figsize=(16, 6))


m.subplot(1, 2, 1)
m.barh(youtube_data.index, youtube_data['Views'], color='skyblue')
m.xlabel('Views')
m.title('Top 10 Artists on YouTube')

m.subplot(1, 2, 2)
m.barh(spotify_data.index, spotify_data['Stream'], color='salmon')
m.xlabel('Streams')
m.title('Top 10 Artists on Spotify')

m.tight_layout()
m.show()
```

## The data for both spotify and youtube showed mostly different artists between them, with only Ed Sheeran and Coldplay being in common. Ed sheeran, in particular, performed at least at the top 2 of each platform.

```python
top_views = df[['Track', 'Views']].nlargest(10, 'Views')
top_likes = df[['Track', 'Likes']].nlargest(10, 'Likes')
top_comments = df[['Track', 'Comments']].nlargest(10, 'Comments')

m.rcParams['font.family'] = 'Arial Unicode MS'
m.figure(figsize=(18, 6))

m.subplot(1, 3, 1)
m.barh(top_views['Track'], top_views['Views'], color='skyblue')
m.xlabel('Views')
m.title('Top 10 Songs with Most Views')

m.subplot(1, 3, 2)
m.barh(top_likes['Track'], top_likes['Likes'], color='salmon')
m.xlabel('Likes')
m.title('Top 10 Songs with Most Likes')

m.subplot(1, 3, 3)
m.barh(top_comments['Track'], top_comments['Comments'], color='lightgreen')
m.xlabel('Comments')
m.title('Top 10 Songs with Most Comments')

m.tight_layout()
m.show()
```

## The graphs show some correlation by having some tracks recurring between them. Despacito is interesting as it has the most in both likes and views while being near at the bottom in the top 10 in comments. The same can be said for Dynamite but for views instead (it was not even part of the top 10).

```python
top_views_info = pd.merge(top_views, df[['Track', 'Artist', 'Energy', 'Danceability']], on='Track', how='left')
top_likes_info = pd.merge(top_likes, df[['Track', 'Artist', 'Energy', 'Danceability']], on='Track', how='left')
top_comments_info = pd.merge(top_comments, df[['Track', 'Artist', 'Energy', 'Danceability']], on='Track', how='left')

print("Top 10 Songs with Most Views:")
print(top_views_info[['Track', 'Artist', 'Energy', 'Danceability']])

print("\nTop 10 Songs with Most Likes:")
print(top_likes_info[['Track', 'Artist', 'Energy', 'Danceability']])

print("\nTop 10 Songs with Most Comments:")
print(top_comments_info[['Track', 'Artist', 'Energy', 'Danceability']])
```

```python
nddf = df.drop_duplicates(subset=['Track'], keep = 'last')
top_songs = nddf.nlargest(10, 'Stream')
top_songs_info = top_songs[['Track', 'Energy', 'Danceability', 'Acousticness']]
combined_labels = top_songs['Track'] + '\n' + top_songs['Artist']

m.figure(figsize=(12, 8))
bar_positions = range(len(top_songs_info))
bar_width = 0.25


m.bar(bar_positions, top_songs_info['Energy'], color='skyblue', width=bar_width, label='Energy')
m.bar([pos + bar_width for pos in bar_positions], top_songs_info['Danceability'], color='salmon', width=bar_width)
m.bar([pos + 2 * bar_width for pos in bar_positions], top_songs_info['Acousticness'], color='lightgreen', width=bar_width)

m.xlabel('Song')
m.ylabel('Score')
m.title('Top 10 Songs by Stream with Energy, Danceability, and Acousticness')
m.xticks([pos + bar_width for pos in bar_positions], combined_labels, rotation=45, ha='right')
m.legend(['Energy', 'Danceability', 'Acousticness'])

m.tight_layout()
m.show()
```

## Remark: The top 10 tracks exhibited high scores in both Energy and Danceability. Half of the tracks exhibits little to no accousticness, 2 are near the mean and there is an outlier (Someone You Loved - Lewis Capaldi).

```python
artist_likes = df.groupby('Artist')['Likes'].sum().reset_index()
artist_likes_sorted = artist_likes.sort_values(by='Likes', ascending=False)
top_10_artists = artist_likes_sorted.head(10)

m.figure(figsize=(12, 8))
m.bar(top_10_artists['Artist'], top_10_artists['Likes'], color='skyblue')
m.xlabel('Artist')
m.ylabel('Total Likes')
m.title('Top 10 Artists by Total Likes')
m.xticks(rotation=45, ha='right')
m.tight_layout()
m.show()
```

## The top2 artists display a significant gap between them and the rest of the list.

```python
single_releases = df[df['Album_type'] == 'single']
best_performing_singles = single_releases.loc[single_releases.groupby('Artist')['Likes'].idxmax()].sort_values(by='Views', ascending=False)

print(best_performing_singles[['Artist', 'Track', 'Likes', 'Views']])
```

## There are 1298 artists who released at least 1 single and out of all of them Charlie Puth and Wiz Khalifa has the best performing ones.

```python
compilation_releases = df[df['Album_type'] == 'compilation']
best_performing_compilation = compilation_releases.loc[single_releases.groupby('Artist')['Likes'].idxmax()].sort_values(by='Views', ascending=False)

print(best_performing_compilation[['Artist', 'Track', 'Likes', 'Views']])
```

## There are 418 artists who released at least 1 compilation and out of all of them, Shakira has the best performing one

```python
top_channels = df.groupby('Channel')['Views'].sum().nlargest(10).sort_values()

m.figure(figsize=(10, 6))
top_channels.plot(kind='barh', color='skyblue')
m.xlabel('Total Views')
m.ylabel('Channel')
m.title('Top 10 Channels Based on Views')
m.gca().invert_yaxis()  
m.show()
```
