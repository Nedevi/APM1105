```python
import requests
import random

def download_words(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, 'wb') as f:
            f.write(response.content)
        print('Dataset saved to file.')
    else:
        print('Failed to download dataset.')

def read_words(filename):
    with open(filename, 'r') as f:
        words = f.readlines()
    return [word.strip() for word in words]

def select_word(words):
    return random.choice(words)

def display_word(word, guessed_letters):
    display = ''
    for letter in word:
        if letter in guessed_letters:
            display += letter
        else:
            display += '_ '
    return display

def get_guess(guessed_letters):
    while True:
        guess = input('Enter a letter: ').lower()
        if len(guess) != 1 or not guess.isalpha():
            print('Please enter a single letter.')
        elif guess in guessed_letters:
            print('You already guessed that letter.')
        else:
            return guess

def main():
    url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'
    filename = 'words.txt'
    download_words(url, filename)
    words = read_words(filename)
    word_to_guess = select_word(words)
    guessed_letters = []
    max_guesses = 10
    guesses_left = max_guesses

    
    while True:
        display = display_word(word_to_guess, guessed_letters)
        print('Word:', display)
        print('Guesses left:', guesses_left)

        if '_' not in display:
            print('Congratulations! You guessed the word:', word_to_guess)
            break
        elif guesses_left == 0:
            print('Sorry, you have run out of guesses. The word was:', word_to_guess)
            break

        guess = get_guess(guessed_letters)
        guessed_letters.append(guess)

        if guess in word_to_guess:
            print('Correct guess!')
        else:
            print('Incorrect guess!')
            guesses_left -= 1
            print('Guesses left:', guesses_left)


main()
```
